```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('words.csv')
```

```python
df['len'] = df['words'].apply(len)
df['chars'] = df['words'].apply(lambda x : list(x))
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(5, 5)
sns.histplot(df, x='len', kde=True, bins=np.arange(1, df['len'].max(), 1), ax=ax)
plt.xlabel("Len of words")
plt.ylabel("Count")
plt.show()
```
